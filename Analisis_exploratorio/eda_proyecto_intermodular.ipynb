# Variables categoricas clientes respecto a abandono 




Jubilados 

- Muestra menos abandono los jubilados ya que son menos  pero los no no jubilados tienen menos abandono en relacion al numero de personas. 


Pareja vs Abandono

-   Los que tienen pareja abandonan menos.

Dependientes vs Abandono

-   Clientes sin dependientes abandonan mucho más que los que tienen dependientes.





```python
import seaborn as sns
import matplotlib.pyplot as plt

# Lista de columnas categóricas
variables = [ 'Jubilado', 'Pareja', 'Dependientes']

# Ajustes de layout
n_cols = 3
n_rows = (len(variables) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
axes = axes.flatten()

# Graficar cada variable
for i, var in enumerate(variables):
    sns.countplot(data=df, x=var, hue='Abandono', ax=axes[i])
    axes[i].set_title(f'{var} vs Abandono')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Conteo')
    axes[i].tick_params(axis='x', rotation=20)

# Eliminar ejes vacíos si hay
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
